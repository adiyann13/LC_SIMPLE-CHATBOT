```python
import os 
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv('GRQO_API_KEY')
```

```python
from langchain_groq import ChatGroq
groq_model = ChatGroq(model ='meta-llama/llama-4-scout-17b-16e-instruct', api_key=groq_api_key)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

groq_model.invoke([
    HumanMessage(content = 'Hoi im adiyann im a data scientist')
])
```

```python
from langchain_core.messages import AIMessage
groq_model.invoke([
    HumanMessage(content='Hie im Adiyann a dat scientist'),
    AIMessage(content ="Hello Adi! Nice to meet you! As a data scientist, you're likely working with data to extract insights and make informed decisions. What kind of projects are you currently working on?"),
    HumanMessage(content='Hey whats my namae nad what do i do?')
])
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables import RunnableWithMessageHistory

store={}
def get_session(sess_id:str)->BaseChatMessageHistory:
    if sess_id not in store:
        store[sess_id]=ChatMessageHistory()
    return store[sess_id]

with_messgae_histry = RunnableWithMessageHistory(groq_model,get_session)
```

```python
config= {'configurable':{"session_id":"chat1"}}
```

```python
with_messgae_histry.invoke([
    HumanMessage(content='Hey im Adiyann a data scientist')],
    config = config
)
```

```python
with_messgae_histry.invoke([
    HumanMessage(content='Hey whats my name ')],
    config = config
)
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
prompt = ChatPromptTemplate.from_messages([
    ("system","You are a helpful ai bot answer all the questions with best of your knowledge"),
    MessagesPlaceholder(variable_name="messages")

])


chain = prompt|groq_model
```

```python
chain.invoke({'messages':[HumanMessage(content ='hi im adiyann')]})
```

```python
with_messgae_histry = RunnableWithMessageHistory(chain, get_session)
```
